STEMMING - It is the text processing technique that is used to extract the root word from the text word . It is used to remove the suffixes from the text word . The main purpose of stemming is to reduce the dimensionality of the text data and to improve the accuracy of the text classification model . example: running -> run , running -> run , running -> run , running -> run , running -> run 

TYPES Of Stemming :
1. Porter Stemming Algorithm: This algorithm is based on the Porter Stemming Algorithm, which is a popular and widely used stemming algorithm . It removes common suffixes from words to reduce them to their base form. Basically , it uses set of rules for removing suffix.

```python
#potter stemming 
from nltk.stem import PorterStemmer 
ps=PorterStemmer()
ListOfWords=['running','flies','easily','history']
for word in ListOfWords:
    print(word +"---------->"+ ps.stem(word))
```

2. lachester Stemmer : This stemming is more agrresive stemming as compare to the PorterStemmer . basically it will reduce the word more shorter.


```python
from nltk.stem import LancasterStemmer
ls=LancasterStemmer()
for word in ListOfWords:
    print(word +"---------->"+ ls.stem(word))
```

SnowBall Stemming: this is more refined and improved version of Porter Stemming . This is gives the support of multi languages 

```python
from nltk.stem import SnowballStemmer
sn=SnowballStemmer("english")
for word in ListOfWords:
    print(word+"------->"+sn.stem(word))
```

RegExp stemmer: this stemmer allows us to make custome made rules means the user can make rules according to itself

```python
from nltk.stem import RegexpStemmer

# Create a stemmer that removes common suffixes like 'ing', 'ly', 'ed', 's'
regexp_stemmer = RegexpStemmer('ing$|ly$|ed$|s$', min=4)

ListOfWords = ['running', 'flies', 'easily', 'history', 'played', 'plays', 'jumps']

for word in ListOfWords:
    print(word + " ----------> " + regexp_stemmer.stem(word))
```

Lemmmetization- the lemmetizaTION IS SIMILAR TO thestemming process but it is more sophesticated than stemming , as it reduced the word into their root form and then identify their parts of speech likie noun , verbs , adjective , adverb of each word or letter.

the output that we get from the lemmetization as root word is called the Lemma , and the output that we get from the stemming called uis called root stem .Lemmetization is more complex than stemmer.
- More Accurate than stemmer
- lemmetization is  used in various nlp tasks like classification , sentiment analysis  and information retrival
- The NLTK provides WordNet Lemmetizer class which is thin wrapper around the wordnet corpus.


```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
lemma=WordNetLemmatizer()
nltk.download("wordnet")

words = ['running', 'flies', 'better', 'easily', 'history', 'ate', 'children']
for word in words:
    print(f"{word} ----------> {lemma.lemmatize(word)}")
```

| Tag | Meaning   |

| `v` | Verb      |

```python
# Adding POS to get better results
print(lemma.lemmatize("running", pos="v"))  # run
print(lemma.lemmatize("better", pos="a"))   # good
print(lemma.lemmatize("ate", pos="v"))      # eat
```

BAg of words: What is Bag of Words (BoW)?

Bag of Words is one of the simplest and most common text representation techniques in Natural Language Processing (NLP).

It represents text (sentences, documents) as a collection (a “bag”) of the words it contains, ignoring grammar and word order but keeping multiplicity (frequency).

🔹 How it works

Create a Vocabulary

Collect all unique words from your dataset.

Example:
Documents:

D1: "I love NLP"

D2: "I love AI"
Vocabulary = {I, love, NLP, AI}

Vectorize Each Document

Each document is turned into a vector of word counts (or presence/absence).

| Word | I | love | NLP | AI |
| ---- | - | ---- | --- | -- |
| Doc1 | 1 | 1    | 1   | 0  |
| Doc2 | 1 | 1    | 0   | 1  |

Variants of BoW

Count Vectorizer → counts occurrences of words.

Binary Vectorizer → just marks presence (0/1).

TF-IDF (Term Frequency – Inverse Document Frequency) → weighs words based on frequency and importance.

🔹 Pros

✅ Simple and easy to implement
✅ Works well for small/medium datasets
✅ Forms the basis for more advanced NLP techniques

🔹 Cons

❌ Ignores word order (so "dog bites man" = "man bites dog")
❌ Leads to high-dimensional sparse vectors for large vocabularies
❌ Cannot capture semantic meaning (similar words are treated differently)

```python
from sklearn.feature_extraction.text import CountVectorizer

# Sample documents
docs = ["I love NLP", "I love AI"]

# Create BoW model
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(docs)

# Vocabulary
print("Vocabulary:", vectorizer.get_feature_names_out())

# Document-Term Matrix
print("Document-Term Matrix:\n", X.toarray())
```
